```python
import pandas as pd
import glob

path = r'/Users/steve_wortmann/Desktop/myFiles' # use your path
all_files = glob.glob(path + "/*.csv")

li = []

for filename in all_files:
    df = pd.read_csv(filename, index_col='id', header=0)
    li.append(df)

frame = pd.concat(li, axis=0, ignore_index=True)[['datetime','lastname','email','lat','long']]
frame['datetime']= pd.to_datetime(frame['datetime'])
```
